### wide_to_long()

#### Example 1 - the suffix is a string

```python
random.seed(10)
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
data_a = pd.DataFrame({
    'Name' : ['Jack', 'Owen', 'Theodore', 'Aiden', 'Samuel', 'Joseph', 'John', 'David', 'Wyatt', 'Matthew'],
    'games_jan': random.sample(range(5, 15), 10),
    'games_feb': random.sample(range(5, 15), 10),
    'scores_jan' : random.sample(range(1, 50), 10),
    'scores_feb' : random.sample(range(1, 50), 10),
})
data_a
```

```python
pd.wide_to_long(data_a, stubnames=['games','scores'], i='Name',j='Month', suffix=r'\w+', sep='_')
```

```python
random.seed(10)
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
data_b = pd.DataFrame({
    'Name' : ['Jack', 'Owen', 'Theodore', 'Aiden', 'Samuel', 'Joseph', 'John', 'David', 'Wyatt', 'Matthew'],
    'Team' : ['team_a', 'team_b', 'team_c']*3 +['team_a'],
    'Games' : random.sample(range(5, 34), 10),
    'day_january': [random.choice(days) for _ in range(10)],
    'day_february': [random.choice(days) for _ in range(10)],
    'day_march': [random.choice(days) for _ in range(10)],
    'scores_january' : random.sample(range(1, 50), 10),
    'scores_february' : random.sample(range(1, 50), 10),
    'scores_march' : random.sample(range(1, 50), 10),
    
#     'scores_1' : random.sample(range(1, 50), 10),
#     'scores_2' : random.sample(range(1, 50), 10),
})
```

```python
data_b
```

Separate the best-days and scores by month

```python
pd.wide_to_long(data_b, 
                stubnames=['day','scores'], 
                i=['Name','Team','Games'], 
                j='Month', 
                sep='_', 
                suffix=r'\w+').reset_index()
```

```python
random.seed(10)
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
data_c = pd.DataFrame({
    'Name' : ['Jack', 'Owen', 'Theodore', 'Aiden', 'Samuel', 'Joseph', 'John', 'David', 'Wyatt', 'Matthew'],
    'Team' : ['team_a', 'team_b', 'team_c']*3 +['team_a'],
    'Games' : random.sample(range(5, 34), 10),
    'day_season1': [random.choice(days) for _ in range(10)],
    'day_season2': [random.choice(days) for _ in range(10)],
    'day_season3': [random.choice(days) for _ in range(10)]
    
#     'scores_1' : random.sample(range(1, 50), 10),
#     'scores_2' : random.sample(range(1, 50), 10),
})
```

```python
data_c
```

Separate the best-days by season.

```python
def rename_col(col_name):
    col_name = col_name.replace('season','')
    if 'day' in col_name:
        col_name = 'best_{}'.format(col_name)
    return col_name
```

```python
data_c = data_c.rename(columns=rename_col)
data_c
```

```python
pd.wide_to_long(data_c, stubnames=['best_day'], i = ['Name','Team','Games'], j='Season', sep='_').reset_index()
```

#### Example: wide_to_long then melt

```python
df = pd.DataFrame(list(zip(
    [0,1,2,3,4],
    [51.62,51.62,51.62,51.62,51.62],
    [-63.81,-63.80,-63.80,-63.80,-63.80],
    [-.04,-.05,-.05,-.06,-.05],
    [0.08,0.09,0.08,0.08,0.09],
    [.1,.1,.07,.11,.11],
    [0.13,.13,.12,.14,.16]
)))
y = [('','FID'),("",'Lat'),("",'Lon'),('PSC','23-May'),("PSC",'18-May'),('Intel','25-May'), ('Intel','28-May')]
col_list = pd.MultiIndex.from_tuples(y)
df.columns = col_list
df
```

```python
df.columns
```

```python
#List comprehension.
#every column has 2 names on different levels.
#return names joined
df.columns = [f'{j}{i}' for i,j in df.columns]
df
```

```python
df_wtl = pd.wide_to_long(df, 
                stubnames = ['23-May','18-May','25-May','28-May'], 
                i=['FID','Lat','Lon'], 
                j='SAT', 
                suffix= '.+', 
                sep='')
df_wtl
```

```python
#reset the index first
#drop Nans
#df_wtl = df_wtl.reset_index()

df_wtl.reset_index().melt(id_vars =['FID','Lat','Lon','SAT'],
            var_name = 'Date',
            value_name = 'NVAL').dropna().reset_index(drop=True)
```

```python
import pandas as pd
codes = ["one","two","three"];
colours = ["black", "white"];
textures = ["soft", "hard"];
N= 100 # length of the dataframe
df = pd.DataFrame({ 'id' : range(1,N+1),
                    'weeks_elapsed' : [random.choice(range(1,25)) for i in range(1,N+1)],
                    'code' : [random.choice(codes) for i in range(1,N+1)],
                    'colour': [random.choice(colours) for i in range(1,N+1)],
                    'texture': [random.choice(textures) for i in range(1,N+1)],
                    'size': [random.randint(1,100) for i in range(1,N+1)],
                    'scaled_size': [random.randint(100,1000) for i in range(1,N+1)]
                   },  columns= ['id', 'weeks_elapsed', 'code','colour', 'texture', 'size', 'scaled_size'])
grouped = df.groupby(['code', 'colour']).agg( {'size': [np.sum, np.average, np.size, pd.Series.idxmax],'scaled_size': [np.sum, np.average, np.size, pd.Series.idxmax]}).reset_index()
```
