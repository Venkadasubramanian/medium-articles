### Cosine curve dataset simulation

Import libraries and set display options

```python
import pandas as pd
pd.set_option('display.max_columns', 100)

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')
```

Create the x and y values

```python
x = np.linspace(0, 2*np.pi, 100)

np.random.seed(321)
noise = np.random.normal(0, 0.5, 100)

y = np.cos(x) + noise
```

```python
y
```

```python
print(x[:10])
print(x[-10:])
```

Save x and y into a dataframe. The dataframe is made up of 2 columns, x and y, each with 100 values

```python
df = pd.DataFrame({'x':x, 'y':y})
df.shape
```

The shape shows the dataframe has 100 rows and 2 columns

```python
df.head()
```

A line plot of x values on the x-axis and the y values(without the noise) on the y-axis, which plots the true cosine curve.

```python
plt.plot(df.x, np.cos(x), color='k')
plt.show()
```

Below is a scatter plot of x values and y values (with noise), and  line plot of x values and y values (without the noise), same curve displayed on the plot above.

```python
plt.scatter(df.x, df.y)
plt.plot(df.x, np.cos(x), color='k')
plt.show()
```

Save the dataframe

```python
df.to_csv('cosine_df.csv', index=False)
```

Modify our dataframe above,by adding 99 new columns, each containing 100 random values which are in no way related to the target column y, or the useful input feature x.

```python
suspect_df = df.copy()

np.random.seed(555)

for n in range(1,100):
    suspect_df['s{}'.format(n)] = np.random.uniform(0, 2*np.pi, 100)

suspect_df.head()
```

```python
suspect_df.to_csv('cosine_df_extra_columns.csv',index=False)
```
